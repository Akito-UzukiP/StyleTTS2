```python
import os
path = "./Data/shimamura_uzuki"
with open(os.path.join(path, "text.txt"),"r") as f:
    lines = f.readlines()
    lines = [line.split("|") for line in lines]
with open(os.path.join(path, "esd.list"),"w") as f:
    for line in lines:
        print(line)
        #filepath = "Data/shimamura_uzuki/{}".format(line[0])
        filepath = line[0]
        f.write("|".join([filepath,"shimamura_uzuki","JP",line[2]]))
```
